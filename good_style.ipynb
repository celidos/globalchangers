```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import elnet_class as ec
```

```python
data = pd.read_csv('august_2h_500 (2).csv', index_col='ID')
word_data = pd.read_csv('august_2h_100_words.csv')
```

```python
num = pd.read_csv('number.csv', index_col='ID')
num.head()
```

```python
data = data.apply(lambda col: col*num['number'], axis=0)
```

```python
data.head()
```

```python
model = ec.ElasticNetPredictor(alpha=0.0001, fit_intercept=False, normalize=False, l1_ratio=0.6, max_iter=1000, 
                               tol=0.001, selection='random')
```

```python
formater = ec.Formater(model=model, row_per_day=12, previous_days=4, previous_words=1, words_to_perform=4000, 
                       collinear_threshold=0.57)
```

```python
%%time
formater.fit(data, word_data)
```

```python
formater.predict(days_to_predict=2)
```

```python
model.process_data(documents_data=data, words_data=word_data, previous_periods=48, previous_words=12, 
                   word_sample_size=4000, collinear_threshhold=0.57)
```
